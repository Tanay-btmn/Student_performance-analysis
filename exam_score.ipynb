```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data_root="/content/student_habits_performance.csv"
perform = pd.read_csv(data_root)
X = perform[["study_hours_per_day","sleep_hours"]].values
Y = perform[["exam_score"]].values
```

```python
perform.plot(kind='scatter', grid = True,
             x = "study_hours_per_day", y ="exam_score")
plt.axis([0, 10, 0, 110])
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
plt.scatter(X[:, 0], Y, color='blue', label='Actual Data')
plt.plot(X[:, 0], model.predict(X), color='red', linewidth=2, label='Linear Fit')
plt.xlabel("Study Hours per Day")
plt.ylabel("Exam Score")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```
